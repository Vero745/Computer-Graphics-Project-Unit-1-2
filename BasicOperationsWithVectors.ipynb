# Algebra lineal

>_**Nota:**_ Debemos considerar (al menos) tres formas diferentes de pensar en los vectores: un vector como una matriz de números (desde la perspectiva de las ciencias de la computación), un vector como una flecha con una dirección y una magnitud (desde la perspectiva de la física), y un vector como un objeto que obedece a la suma y la escala (desde la perspectiva de la matemática).

El álgebra lineal es el estudio de vectores y ciertas reglas de álgebra para manipular vectores. En general, los vectores son objetos especiales que se pueden sumar y multiplicar por escalares para producir otro objeto del mismo tipo. Desde un punto de vista matemático abstracto, cualquier objeto que satisfaga estas dos propiedades puede considerarse un vector. A continuación, se muestran algunos ejemplos de tales objetos vectoriales: 

1. **Vectores geométricos:** Los vectores geométricos son segmentos dirigidos, que se pueden dibujar (al menos en dos dimensiones). Se pueden sumar dos vectores geométricos $\overrightarrow{x}, \: \overrightarrow{y}$, de modo que $\overrightarrow{x}+ \: \overrightarrow{y} = \overrightarrow{z}$ es otro vector geométrico. Además, la multiplicación por un escalar λ $\overrightarrow{x}$, λ ∈ R, también es un vector geométrico. De hecho, es el vector original escalado por λ. Por lo tanto, los vectores geométricos son instancias de los conceptos vectoriales introducidos anteriormente. Interpretar los vectores como vectores geométricos nos permite usar nuestras intuiciones sobre la dirección y la magnitud para razonar sobre operaciones matemáticas.


```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt

def graficar(vals, cols, alpha = 1):
    
    plt.figure()
    plt.axvline(x = 0, color = "grey", zorder = 0)
    plt.axhline(y = 0, color = "grey", zorder = 0)
    
    for i in range(len(vals)):
        x = np.concatenate([[0,0], vals[i]])
        plt.quiver([x[0]],
                   [x[1]],
                   [x[2]],
                   [x[3]],
                  angles='xy', scale_units='xy', scale=1,
                   color=cols[i],
                  alpha = alpha)

v1 = np.array([2,5])
v2 = np.array([3,2])
```

# Suma 

```python
v1v2 = v1+v2

graficar([v1, v2, v1v2],["black","red","blue"])
plt.xlim(-1, 8)
plt.ylim(-1, 15)
```

# Resta

```python
v1v2 = v1-v2

graficar([v1, v2, v1v2],["black","red","blue"])
plt.xlim(-2, 7)
plt.ylim(-1, 15)
```

 # Multiplicación

```python
v1v2 = v1*v2

graficar([v1, v2, v1v2],["black","red","blue"])
plt.xlim(-2, 7)
plt.ylim(-1, 15)
```
